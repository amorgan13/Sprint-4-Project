```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
import os
print(os.getcwd())
```

```python
# Load dataset
df = pd.read_csv('vehicles_us.csv')
```

```python
# First few rows
print(df.head())
```

```python
# Shape of the data
print(df.shape)
```

```python
# Data types and non-null counts
print(df.info())
```

```python
# Statistical summary
print(df.describe())
```

```python
# Missing values
print(df.isnull().sum())
```

```python
# Histogram 1: distribution of Paint Color
fig1 = px.histogram(df, x="paint_color", title="Distribution of Paint Color")
fig1.show()
```

```python
# Histogram 2: distribution of price
fig2 = px.histogram(df, x="price", title="Distribution of Price")
fig2.show()
```

```python
# Scatterplot 1: model vs model year
fig3 = px.scatter(df, x="model_year", y="model", title="Model vs Model Year")
fig3.show()
```

```python
# Scatterplot 2: Model Year vs Price
fig4 = px.scatter(df, x="model_year", y="price", title="Model Year vs Price")
fig4.show()
```
